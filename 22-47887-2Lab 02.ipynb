# for statement
<p style="text-align: justify">The for statement in Python differs a bit from what you may be used to in C or C++. Rather than always giving the user the ability to define both the iteration step and halting condition, Python’s for statement iterates over the items of any sequence (a list or a string), in the order that they appear in the sequence.</p>

```python
#Basic for Loop
for i in range(5,10):
    print("Iteration:", i)
```

```python
#basic for loop
for f in range(6,20):
  print ("iteration :", f)
```

<b>Exercise 3:</b> Define a <b>method/function</b> which will take two lists of numbers and returns a list (sum of two lists)
<p>Example:<br>
    Input: [3,4,5,1] and [6,7,2,8]<br>
    output: [9,11,7,9]</p>


```python
# code here
def sum_of_two_lists(list1, list2):
    # Check if both lists are the same length
    if len(list1) != len(list2):
        return "Error: Lists must be of the same length"

    # Initialize an empty list to store the result
    result = []

    # Loop through both lists and sum corresponding elements
    for i in range(len(list1)):
        result.append(list1[i] + list2[i])

    return result

# Example usage
list1 = [3, 4, 5, 1]
list2 = [6, 7, 2, 8]

# Call the function
sum_list = sum_of_two_lists(list1, list2)

# Output the result
print("Sum of the two lists:", sum_list)
```

<b> Exercise 4:</b> Write a <b>method</b> which will take the <b>full name</b> as input and return both the first name and last name.
<p>Example:<br>
    Input: Arafat Rahman Sunny<br>
    Output: First name: Arafat, Last name: Sunny</p>

```python
# code here
def extract_first_last_name(full_name):
    # Split the full name by spaces
    parts = full_name.strip().split()

    # Check if there are at least two parts
    if len(parts) < 2:
        return "Please enter at least a first and last name."

    first_name = parts[0]
    last_name = parts[-1]  # Take the last part as last name

    return f"First name: {first_name}, Last name: {last_name}"

# Example usage
name = "Arafat Rahman Sunny"
result = extract_first_last_name(name)
print(result)
```
